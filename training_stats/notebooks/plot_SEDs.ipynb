```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from matplotlib import cm
from matplotlib.colors import to_hex

from matplotlib.colors import Normalize
from matplotlib.cm import get_cmap
from matplotlib.colors import to_hex


matplotlib.rc('ytick', labelsize=12)
matplotlib.rc('xtick', labelsize=12)
matplotlib.rcParams.update({'font.size': 8})


base_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))

SED_folder_O = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/O')
SED_folder_B = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/B')
SED_folder_A = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/A')
SED_folder_F = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/F')
SED_folder_G = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/G')
SED_folder_K = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/K')
SED_folder_M = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/M')


SED_folder_WK = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/WK')
SED_folder_RK = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/RK')

def plot_seds_colormap_by_value(folder, list_file=None, cmap_name='plasma', cmin=0, cmax=None):

    # Read sed list and delete cache folder path
    if list_file is not None:
        with open(list_file, 'r') as f:
            valid_files = [line.split()[0].replace("PICKLES/", "") for line in f if line.strip()]

        # List existing files in folders
        files = [f for f in os.listdir(folder) if f in valid_files]

        if not files:
            print(f"No valid SED files found in {folder} with {list_file}")
            return

        vals = []

    # tape 1 : rcuprer les valeurs (ici, df.T[1][0] = valeur caractristique du SED)
    for SED in files:
        file = os.path.join(folder, f'{SED}')
        df = np.loadtxt(file)
        vals.append(df.T[1][df.T[0]==3850])

    vals = np.array(vals)

    # tape 2 : normalisation personnalise
    vmin = cmin if cmin is not None else vals.min()
    vmax = cmax if cmax is not None else vals.max()
    norm = Normalize(vmin=vmin, vmax=vmax)
    cmap = get_cmap(cmap_name)

    # tape 3 : trac avec couleur normalise
    for SED, val in zip(files, vals):
        file = os.path.join(folder, f'{SED}')
        df = np.loadtxt(file)
        color = to_hex(cmap(norm(val)))
        plt.loglog(df.T[0], df.T[1], label=SED, color=color)



plt.figure(figsize=(8, 5))

list_file = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/lib_bin/DES_STAR.list')

plot_seds_colormap_by_value(SED_folder_O, list_file, 'Blues')
plot_seds_colormap_by_value(SED_folder_B, list_file, 'Purples')
plot_seds_colormap_by_value(SED_folder_A, list_file, 'Greys')
plot_seds_colormap_by_value(SED_folder_F, list_file, 'Reds')
plot_seds_colormap_by_value(SED_folder_G, list_file, 'Greens')
plot_seds_colormap_by_value(SED_folder_K, list_file, 'Oranges')
plot_seds_colormap_by_value(SED_folder_M, list_file, 'RdPu')

plt.xlabel("Wavelength (A)")
plt.axvspan(3500, 11000, color='gray', alpha=0.1, label='SDSS filter region')
plt.xlim(3450, 11050)
plt.ylim(1e-2, 10)
plt.ylabel("Transmission (Normed at 550A)")
plt.tight_layout()
# plt.legend()
plt.show()
```
