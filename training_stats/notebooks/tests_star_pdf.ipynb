# star pdf output tests

We developed an equivalent of the PDZ from zphota but for stars, PDF_star. We considered the transition from O-type to M-type stars as continuous, hence the better the fit is to the proper star spectral-type, the smaller the CHI2-fit is, thus the higher the PDF is. Resulting in modes, std, tail mass... of the resulting PDF. This notebook stands to test those assumption on the DES calibrated star catalog.  

## Initialization and output

```python
%matplotlib widget
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
from scipy.integrate import trapezoid
from scipy.optimize import curve_fit
```

```python
# .prob path
base_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
# pdf_path = os.path.join(base_dir, 'simulation_catalogs/DES/DES_STARCAT/WORK_COMPLETE2/zphota/star_pdfs/DESstar_PICKLESfull_DES_10000_PDFstar.prob')
pdf_path = os.path.join(base_dir, 'simulation_catalogs/star_gal/PDFs/DC1_PICKLES_LSST_10000_PDFstar.prob')

pdf_file = np.loadtxt(pdf_path)
```

first plots

```python
rows=np.linspace(1,9900,10, dtype='int')
pdz_rows = [pdf_file[row][1:] for row in rows]

# Plot
plt.figure(figsize=(8, 5))

for (row,i) in zip(pdz_rows,rows):
    x=np.linspace(1,len(row),len(row))
    plt.plot(x,row, label=f'{pdf_file[i][0]}')

# plt.title(f'PDF_star for IDENT={pdf_file[row][0]}')
plt.xlabel('Spectral type (ST)')
plt.ylabel('P(ST)')
plt.legend()

plt.tight_layout()
plt.show()
```
